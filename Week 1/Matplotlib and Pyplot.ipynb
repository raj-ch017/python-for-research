# Introduction to Matplotlib and Pyplot
---

**Matplotlib** is a widely used Python library for creating high-quality static, animated, and interactive visualizations in Python. It offers an array of customization options for a variety of plot types, including line plots, scatter plots, bar plots, and histograms, among others. 

**Pyplot**, a submodule of the Matplotlib library, provides a collection of functions that makes creating visualizations with Matplotlib even more straightforward. It is particularly useful for interactive work, such as examining simulation results or performing exploratory data analysis.

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

The library has been imported and the plotting process can be initiated.

The **plot** function is a helpful command that can be utilized to plot lines and markers. The basic form of plot includes a single argument, the plot function, and every y-axis value that will be plotted against its corresponding index value on the x-axis.

As Python indexing begins at 0, the first element of the array is located at `x=0`.

```python
list1 = [0,1,4,9,16]

plt.plot(list1)
```

Python returns a **matplotlib** object.

```python
plt.plot(list1);
```

Providing a semicolon **;** to the end of the statement only provides the plot.

The `plot` function can be used with two arguments to plot the points in a graph. The first argument provides the x-coordinates of the points, and the second argument provides the corresponding y-coordinates.


Let's create a vector `input_vector`.

```python
input_vector = np.linspace(0,10,50)
```

The code block below uses list comprehension and the `sin` function from the `math` module to create a vector named `sin_values`.

Each element in the `sin_values` vector is obtained by applying the `sin` function on the corresponding element in the `input_vector`.

```python
sin_values = [sin(x_in) for x_in in input_vector]
```

```python
plt.plot(input_vector,sin_values);
```

The plot representes the graph of a **sine** function on the input values [0,10]

The `plot` function can take a third argument, which is a format string used to specify the color, marker, and line type of the plot. This string contains a combination of characters that represent the desired color, marker, and line type. 

For example, `bo--` specifies that the plot should be a blue circle with a dashed line.

```python
cos_values = [cos(x_in) for x_in in input_vector]
```

```python
plt.plot(input_vector,cos_values,"bo--");
```

The `linewidth` parameter is set to 2 to increase the line width, and `markersize` is set to 12 to increase the size of the markers.

```python
plt.plot(input_vector,cos_values,"go-",linewidth=2,markersize=12);
```

There are a few important features that can be added to plots.

    legend() - adding a legend
    axis()  - adjust axes 
    xlabel(), ylabel() - set axis labels
    savefig() - save a figure
     
In the case of saving a plot, the file format extension specifies the format of the file, such as a `pdf` or `png`.     

The method `xlabel` is utilized to label the horizontal axis, while the method `ylabel` is used to label the vertical axis. This helps in providing a clear and comprehensive understanding of the plotted data.

```python
plt.plot(input_vector,sin_values,"r--");
plt.plot(input_vector, cos_values,"b-");
plt.xlabel("$x$")
plt.ylabel("Function of x")
```

The **axis** method can be used to adjust the axis on the plot by providing a list as an argument:

    [xmin,xmax,ymin,ymax]

```python
plt.axis([-0.5,10.5,-1.25,1.25])

plt.plot(input_vector,sin_values,"r--");
plt.plot(input_vector, cos_values,"b-");
plt.xlabel("$x$")
plt.ylabel("Function of x")
```

To add a legend to a plot, labels must first be given to the individual plots. This can be done using the `label` parameter within the plot function. 

The next step is to use the `legend` method to show the legend. The loc parameter can be used to specify the location of the legend on the plot if needed.

```python
plt.axis([-0.5,10.5,-1.25,1.25])

plt.plot(input_vector,sin_values,"r--",label="Sin(x)");
plt.plot(input_vector, cos_values,"b-",label="Cos(x)");
plt.xlabel("$x$")
plt.ylabel("Function of x")
plt.legend();
```

The figure can be saved as a file using the `savefig` method. A string input, containing the name of the file along with its file extension, is passed to the method.

```python
plt.axis([-0.5,10.5,-1.25,1.25])

plt.plot(input_vector,sin_values,"r--",label="Sin(x)");
plt.plot(input_vector, cos_values,"b-",label="Cos(x)");
plt.xlabel("$x$")
plt.ylabel("Function of x")
plt.legend();
plt.savefig("myplot.pdf")
```

### Plotting Using Logarithmic Axis
    
In many scientific and engineering applications, it is often useful to plot data using logarithmic scales. This can help to reveal patterns and relationships that might not be apparent in a linear plot. 

With Matplotlib, it is possible to create plots with logarithmic scales using the `semilogx`, `semilogy`, or `loglog` functions. These functions are similar to the plot function but take logarithmically scaled axes as inputs. By default, the logarithm is taken in base 10, but this can be changed as needed using the base keyword argument.

The **plot** functions that can be used to make logarithmic plots are:

- semilogx() - plots the horizontal axis on a log scale and vertical axis in original scale
- semilogy() - plots the vertical axis on a log scale and horizontal axis in original scale
- loglog() - plots both the axes on logarithmic scale

```python
#recreating the input vector

input_vector = np.linspace(0.1,10,50)

second_power = [(x_in ** 2) for x_in in input_vector]
fifth_power = [(x_in ** 5) for x_in in input_vector]
```

To plot the log of the `input_vector` on the horizontal axis, a small modification to the `input_vector` is required. 

As log(0) is undefined, the start argument of linspace needs to be changed to 0.1 instead of 0.

```python
plt.plot(input_vector,second_power,'g-',label="x^2");
plt.plot(input_vector,fifth_power,'b-',label="x^5");
plt.xlabel("$x$")
plt.ylabel("Function of x")
plt.legend();
```

The plotted functions exhibit distinct characteristics that can be analyzed.


The first function, which is defined as x to the power of 5, produces large y values, which heavily dominate the plot. As a result, the graph increases drastically for larger values of x.


The second function, which is defined as x to the power of 2, appears as a horizontal line in the graph due to its lower magnitude compared to the first function.

```python
plt.loglog(input_vector,second_power,'g-',label="$log$(x^2)");
plt.loglog(input_vector,fifth_power,'b-',label="$log$(x^5)");
plt.xlabel("$x$")
plt.ylabel("Log Function of x")
plt.legend();
```

The `loglog` method is used to plot the logarithmic values of the input vector and corresponding functions. This allows for a linear plot to be created and provides better understanding of x^2 compared to x^5.


The line representing **x^5** as a steeper slope since taking the log converts the power to the slope of the line:

    y = x^5
    log(y) = log(x^5)
    log(y) = 5 * log(x)
    
Since the power of **x^5** is greater than **x^2**, the slope of **log(x^5)** line is steeper than the slope of **log(x^2)**.
    

### Generating Histograms

Histograms are a useful tool in data analysis and visualization. They allow for the graphical representation of the distribution of a dataset by dividing the data into discrete bins and plotting the frequency of the observations that fall within each bin. In Python, histograms can be generated using the `hist` function from the `plot` library.

The code block below generates an array of 1000 random numbers sampled from a normal (Gaussian) distribution using the `np.random.normal()` function from the NumPy library. The size argument specifies the size of the array to be generated.

Since no mean or standard deviation is specified, the default values (0 and 1) are used for the mean and standard deviation of the normal distribution, respectively.

```python
first_sample = np.random.normal(size=1000)
```

```python
plt.hist(first_sample);
```

Setting the **density** argument to **True**, the vertical axis on the histogram plot consists of the proportion of observations that fall in each bin, instead of the number of observations in each bin.

```python
plt.hist(first_sample,density=True);
```

When providing the location of bins in the hist function, the `bins` keyword argument can be used. Equally spaced intervals for the bins can be created using the `linspace` method of the Numpy library.

```python
plt.hist(first_sample,density=True,bins=np.linspace(-5,5,21));
```

The current histogram plot shows a difference in its appearance compared to the previous ones due to the specification of 20 bins between the values of -5 and 5.

#### Gamma Distribution

The gamma distribution is a mathematical concept that is used to model a wide range of phenomena in various fields such as physics, engineering, and finance.

It is a continuous probability density function, meaning that it assigns probabilities to the range of values that a random variable can take on. The gamma distribution is characterized by two parameters, which can be used to adjust its shape and scale. It is a versatile distribution that can take on many different shapes, including skewed or symmetric, unimodal or multimodal, and short-tailed or long-tailed.

#### Plot Subplot 

The method `subplot` enables the creation of multiple subplots within a single figure. It takes in three arguments, with the first two specifying the number of rows and columns in the subplot grid, and the third argument indicating the position of the subplot within the grid.


The plot number always starts at 1. So in a (2,3) subplot, the plot numbers range from 1 to 6, where the plot numbers are incremented rows first.

The code block below generates an array of 100,000 random numbers from a gamma distribution using the NumPy library.

The gamma distribution is characterized by two parameters: `shape parameter` and `scale parameter`.

In this case, the shape parameter is set to 2 and the scale parameter is set to 3, which means that the generated values will have an average of 6 (shape x scale). The higher the shape parameter, the more the distribution is skewed to the right, while the higher the scale parameter, the more spread out the distribution.

```python
gamma_dist = np.random.gamma(2,3,100000)
```

```python
plt.figure()

# plot in location (1,1)
plt.subplot(221)
plt.hist(gamma_dist,bins=30);

# plot in location (1,2)
plt.subplot(222)
plt.hist(gamma_dist,bins=30,density=True);

# plot in location (2,1)
plt.subplot(223)
plt.hist(gamma_dist,bins=30,cumulative=30);

# plot in location (2,2)
plt.subplot(224)
plt.hist(gamma_dist,density=True,cumulative=True, histtype='step');
```

A figure is created using the `figure` method. Afterwards, each histogram is inserted into an individual subplot using the `subplot` method
